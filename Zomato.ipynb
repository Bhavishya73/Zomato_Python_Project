```python
# import library for project

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Create the DataFram and read the data

df = pd.read_csv('Zomato data .csv')
# print(df)
df
```

```python
#  Convert the data type of Rate column(Data cleaning and processing)

def handlerate(value):
    value = str(value).split('/')
    value = value[0]
    return float(value)

df['rate'] = df['rate'].apply(handlerate)
df
```

```python
df.info()
```

```python
# Q1 What type of restaurant do the majority of customers order from?

df.head()
```

```python
sns.countplot(x=df['listed_in(type)'])
plt.xlabel('Type of Restaurant')
plt.ylabel('Percentage')

# Conclusion : - majority of the restaurant fall in dining category
```

```python
# Q2 How many votes has each type of restaurant received from customers?

df.head()
```

```python
grouped_data = df.groupby('listed_in(type)') ['votes'].sum()
result = pd.DataFrame({'votes': grouped_data})
plt.plot(result, c="green", marker="o")
plt.xlabel("Type of restaurant", c="red", size=20)
plt.ylabel("Votes", c="red", size=20)
```

```python
# conclusion - dinning resturants has recieved maximum votes
```

```python
# Q3 What are the ratings that the majority of restaurants have received? 

df.head()
```

```python
plt.hist(df['rate'],bins =5)
plt.title("Rating Distribution")
plt.show()
```

```python
# conclusion  : - the majority restaurent recived the rating from 3.5 to 4
```

```python
# Q4 Zomato has observed that most couples order most of their food online. What is their average spending on each order?

df.head()
```

```python
couple_data = df['approx_cost(for two people)']
sns.countplot(x=couple_data)
```

```python
# conclusion the majority of couples preferr resturants with an approximate cost of 300 rupees
```

```python
# Q5  Which mode (online or offline) has received the maximum rating?

df.head()
```

```python
plt.figure(figsize = (6,6))
sns.boxplot(x='online_order' , y = 'rate' , data = df)
```

```python
# conclusion :- offiline order recived lower rating then online oerder
```

```python
# Q6 Which type of restaurant received more offline orders, so that Zomato can provide those customers with some good offers?

df.head()
```

```python
pivot_table = df.pivot_table(index='listed_in(type)', columns='online_order', aggfunc='size', fill_value=0)
sns.heatmap(pivot_table, annot=True, cmap="YlGnBu", fmt='d')
plt.title("Heatmap")
plt.xlabel("Online Order")
plt.ylabel("Listed In (Type)")
plt.show()
```
